```python
import matplotlib.pyplot as plt
import numpy as np

n_groups = len(df_averages_braking_by_year)

fig, ax = plt.subplots()
bar_width = 0.2

index = np.arange(n_groups)

bar1 = plt.bar(index, df_averages_braking_by_year['average_maxwaarde'], bar_width, label='Average Max g-force recorded during the incident')
bar2 = plt.bar(index + bar_width, df_averages_braking_by_year['average_duration_seconds'], bar_width, label='Average Duration of the incident in seconds')
bar3 = plt.bar(index + 2 * bar_width, df_averages_braking_by_year['average_speed_kmh'], bar_width, label='Average Speed at the start of the incident in kmh')
bar4 = plt.bar(index + 3 * bar_width, df_averages_braking_by_year['end_speed_kmh'], bar_width, label='Average Speed at the end of the incident in kmh')

plt.xlabel('Year')
plt.ylabel('Averages')
plt.title('Averages by Year (braking category)')
plt.xticks(index + 1.5 * bar_width, df_averages_braking_by_year['year'])
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

def add_values(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2.0, height, '%.2f' % height,
                 ha='center', va='bottom')

add_values(bar1)
add_values(bar2)
add_values(bar3)
add_values(bar4)

plt.show()
```

### ANALYZING THE 'incident_severity' FEATURE

After looking at the datasets ('safe_driving', 'precipitation', 'temperature', 'wind', 'greenery', 'breda_road', 'accidents'), we have established that the 'incident_severity' column can potentially be a good target feature for the model.

#### CHECKING HOW MANY DISTINCT TYPES OF SEVERITIES EXTIST IN THE 'safe_driving' TABLE

```python
conn = psycopg2.connect(**db_params)

cursor = conn.cursor()

cursor.execute("""SELECT DISTINCT TRIM(incident_severity) FROM group12_warehouse.safe_driving""")

result = cursor.fetchall()

for row in result:
    print(row)

cursor.close()

conn.close()
```

```python
conn = psycopg2.connect(**db_params)

cursor = conn.cursor()

cursor.execute("""SELECT COUNT(DISTINCT TRIM(incident_severity)) FROM group12_warehouse.safe_driving""")

result = cursor.fetchall()

for row in result:
    print(row)

cursor.close()

conn.close()
```

There are 30 types of distinct sverity types. There are three severity types for the 'ACCELERATING' category, 'HA1', 'HA2' and 'HA3', same for the 'BRAKING' category, 'HB1', 'HB2' and 'HB3'. For the 'HARSH CORNERING' category there are sverity types from 'HC1' to 'HC21', skipping the 'HC9' and 'HC12' severity types. Last but not least, for the 'SPEEDING' category there are five severity types, from 'SP1' to 'SP5'.

If there was to be built a model predicting the 'incident_severity' feature, a CNN would work perfectly since it is optimal for multi-label classification. However, a more conventional model, like Random Trees, would probably perform underwhelmingly since it is not made to classify 30 classes. There is a fix for the convetional model problem, and that would be creating a new numerical feature that mirrors the 'incident_severity' feature, but with numerical values. An example of this would be replacing the 'HC11' severity type with the number 11, as this would show the intensity of the incident in a numerical value. This would be a good idea if there were to be two models, one classifying the 'category', as the severity types are grouped by 'category' ('SP1' severity type for the 'SPEED' category, 'HC11' severity type for the 'HARSH CORNERING' category, etc.), and one predicting the new numerical feature.

#### FEATURES THAT CAN BE USED TO PREDICT 'incident_severity'

```python
import pandas as pd

conn = psycopg2.connect(**db_params)

cursor = conn.cursor()

cursor.execute("""SELECT * FROM group12_warehouse.safe_driving""")

rows = cursor.fetchall()

colnames = [desc[0] for desc in cursor.description]

df_safe_driving = pd.DataFrame(rows, columns=colnames)

cursor.close()

conn.close()
```

Firstly going to see any features that correlate with the 'incident_severity' feature from the 'safe_driving' table

```python
df_safe_driving['incident_severity'] = df_safe_driving['incident_severity'].str.strip()
```

```python
print(df_safe_driving['incident_severity'].value_counts())
```

Going to do it by category, starting with 'HARSH CORNERING'

```python
df_safe_driving_incident_severity_hc = df_safe_driving[df_safe_driving['incident_severity'].str.startswith('HC')]
```

```python
print(df_safe_driving_incident_severity_hc['incident_severity'].value_counts())
```

```python
import matplotlib.pyplot as plt

grouped_df_safe_driving_incident_severity_hc = df_safe_driving_incident_severity_hc[['incident_severity', 'speed_kmh', 'end_speed_kmh']].groupby('incident_severity').mean()

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Incident Severity')
ax1.set_ylabel('Average Speed At The Start Of The Incident', color=color)
grouped_df_safe_driving_incident_severity_hc['speed_kmh'].plot(kind='bar', ax=ax1, color=color, position=0, width=0.4)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Average Speed At The End Of The Incident', color=color)
grouped_df_safe_driving_incident_severity_hc['end_speed_kmh'].plot(kind='bar', ax=ax2, color=color, position=1, width=0.4)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Average Speed At The Start Of The Incident And At The End Of The Incident By Incident Severity')

plt.show()
```

There are some anomalies when it comes to the correlation between the speed at the start and end of the incident and incident severity. The severity type that has the lowest average speed at the start and the end of the incident is 'HC11', which is unsual since, theoretically, that place should be held by 'HC1'. However, this can have something to do with how the severity type was created, 'HC11 = HB3 X HC2'. Maybe the speed features are not the best when it comes to predicting the severities. What is good is that each severity type has a different pattern of speed and no two severity types are the same, some are close to eachother.

```python
import matplotlib.pyplot as plt

grouped_df_safe_driving_incident_severity_hc = df_safe_driving_incident_severity_hc[['incident_severity', 'duration_seconds']].groupby('incident_severity').mean()

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Incident Severity')
ax1.set_ylabel('Average Duration Of The Incident', color=color)
grouped_df_safe_driving_incident_severity_hc['duration_seconds'].plot(kind='bar', ax=ax1, color=color, position=0, width=0.4)
ax1.tick_params(axis='y', labelcolor=color)

plt.title('Average Duration Of The Incident By Incident Severity')

plt.show()
```

The same applies to the duration of the incident. It is still good that there are not too many severity types that have the same patterns, which is good when it comes to predicting it.

```python
import matplotlib.pyplot as plt

grouped_df_safe_driving_incident_severity_hc = df_safe_driving_incident_severity_hc[['incident_severity', 'maxwaarde']].groupby('incident_severity').mean()

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Incident Severity')
ax1.set_ylabel('Average Maximum G-force Recorded During The Incident', color=color)
grouped_df_safe_driving_incident_severity_hc['maxwaarde'].plot(kind='bar', ax=ax1, color=color, position=0, width=0.4)
ax1.tick_params(axis='y', labelcolor=color)

plt.title('Average Maximum G-force Recorded During The Incident By Incident Severity')

plt.show()
```

```python
import pandas as pd

conn = psycopg2.connect(**db_params)

cursor = conn.cursor()

cursor.execute("""CREATE TABLE IF NOT EXISTS group12_warehouse.accident_data_17_23_safe_driving_inner_join AS
                    SELECT a."Year",
                            a."Accident severity",
                            a.municipality, 
                            a.town,
                            a."First Mode of Transport",
                            a."Second mode of Transport",
                            a."Area Type",
                            a."Light condition",
                            a."Road Location",
                            a."Road condition",
                            a."Road surface",
                            a."Road situation",
                            a."Speed limit",
                            a.street,
                            a.weather,
                            a.accidents,
                            AVG(b.longitude) AS avg_longitude,
                            AVG(b.latitude) AS avg_latitude
                    FROM group12_warehouse.accident_data_17_23 a
                    INNER JOIN group12_warehouse.safe_driving b
                    ON a.street = b.road_name
                    GROUP BY a."Year",
                            a."Accident severity",
                            a.municipality, 
                            a.town,
                            a."First Mode of Transport",
                            a."Second mode of Transport",
                            a."Area Type",
                            a."Light condition",
                            a."Road Location",
                            a."Road condition",
                            a."Road surface",
                            a."Road situation",
                            a."Speed limit",
                            a.street,
                            a.weather,
                            a.accidents""")

conn.commit()

cursor.close()

conn.close()

print("Table created succesfully!")
```
